```python
df_true = df[df.fake == 0]
df_fake = df[df.fake == 1]
```

```python
print(df_true.shape[0])
print(df_fake.shape[0])
```

```python
df_class0_under = df_true.sample(df_fake.shape[0])
```

```python
df_final = pd.concat([df_class0_under, df_fake], axis = 0)
```

```python
df_final.fake.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X_train , X_text , y_train, y_test = train_test_split(df_final['text'], df_final['fake'], test_size = 0.2, random_state = 2022, stratify = df_final['fake'])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
clf = Pipeline(
    [('vect', CountVectorizer(ngram_range=(1,3))),
     ('knn', KNeighborsClassifier(n_neighbors = 10, metric = 'euclidean'))
     ]
)
```

```python
from sklearn.metrics import classification_report
clf.fit(X_train, y_train)
y_pred = clf.predict(X_text)
print(classification_report(y_test, y_pred))
```

```python
clf2 = Pipeline(
    [('vect', CountVectorizer(ngram_range=(1,3))),
     ('knn', KNeighborsClassifier(n_neighbors = 10, metric = 'cosine'))
     ]
)
```

```python
clf2.fit(X_train, y_train)
y_pred_2 = clf2.predict(X_text)
print(classification_report(y_test, y_pred_2))
```

```python
from sklearn.ensemble import RandomForestClassifier

clf3 = Pipeline(
    [('vect', CountVectorizer(ngram_range=(3,3))),
     ('rf', RandomForestClassifier())
     ]
)
clf3.fit(X_train, y_train)
y_pred_3 = clf3.predict(X_text)
print(classification_report(y_test, y_pred_3))
```

```python
# random forest is the best !
```

```python
from sklearn.naive_bayes import MultinomialNB

clf4 = Pipeline(
    [('vect', CountVectorizer(ngram_range=(1,2))),
     ('NB', MultinomialNB())
     ]
)
clf4.fit(X_train, y_train)
y_pred_4 = clf4.predict(X_text)
print(classification_report(y_test, y_pred_4))
```
